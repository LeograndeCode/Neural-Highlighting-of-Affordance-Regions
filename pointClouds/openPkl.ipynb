```python
%cd /content/drive/MyDrive/Neural-Highlighting-of-Affordance-Regions/pointClouds/

import pickle
import open3d as o3d
import numpy as np
import gc

'''
with open('full_shape_train_data.pkl', 'rb') as file:
  data = pickle.load(file)

print(type(data))
print(len(data))

if isinstance(data, list):
    point_cloud = data[1]

print(point_cloud)

coordinates = np.array(point_cloud['full_shape']['coordinate'])
coordinates_pushable = np.array(point_cloud['full_shape']['label']['pushable'])

print(coordinates.shape)
print(coordinates_pushable.shape)

# Convert to Open3D PointCloud format
pcd = o3d.geometry.PointCloud()
pcd.points = o3d.utility.Vector3dVector(coordinates)

# Save to a PLY file
o3d.io.write_point_cloud("door2_point_cloud.ply", pcd)
'''

# Load the pickle file
with open('full_shape_train_data.pkl', 'rb') as file:
    data = pickle.load(file)

print(type(data))
print(len(data))

semantic_class = 'Mug'
label_name = 'grasp'

# Check if the data is a list
if isinstance(data, list):
    # Filter the point clouds based on the 'semantic class' attribute
    point_clouds = [
        pc for pc in data
        if pc.get('semantic class') == semantic_class
    ]

print(f"Number of point clouds with 'semantic class' equal to {semantic_class}: {len(point_clouds)}")

# Process each door point cloud
for i, point_cloud in enumerate(point_clouds):
  if(i > 2):
    # Extract coordinates
    print(point_cloud)
    coordinates = np.array(point_cloud['full_shape']['coordinate'])
    print(f"Point cloud {i} shape: {coordinates.shape}")

    # Extract label pushable
    label = np.array(point_cloud['full_shape']['label'][label_name])
    print(f"Label {i} shape: {label.shape}")

    # Convert to Open3D PointCloud format
    pcd = o3d.geometry.PointCloud()
    pcd.points = o3d.utility.Vector3dVector(coordinates)

    # Save the point cloud as a PLY file
    ply_file_name = f"{semantic_class}_point_cloud_{i}.ply"
    o3d.io.write_point_cloud(ply_file_name, pcd)
    print(f"Saved {ply_file_name}")

    # Save the label as a .npy file
    label_file_name = f"{semantic_class}_point_cloud_label_{label_name}_{i}.npy"
    np.save(label_file_name, label)
    print(f"Saved {label_file_name}")

    if(i == 7):
      break

del data
gc.collect()
```
